```python
cars=pd.read_csv('mtcars.csv')
```

```python
cars.loc[2,'hp']=np.nan    #putting some nan values
cars.loc[3,'hp']=np.nan
cars.loc[5,'mpg']=np.nan
cars.loc[6,'mpg']=np.nan
cars.head(6)
```

```python
 #filling NaN value by .....some approaxima....methods
    
cars.fillna(method='ffill')         #filling by forward
cars.fillna(method='bfill')         #filling by backward
cars.fillna(cars.mean())            #filling by mean
cars.dropna(inplace=True)     #drop NaN
cars.dropna(how='all')            #drop only all columns of a raw are NaN
```
